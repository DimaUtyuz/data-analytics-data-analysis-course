```python
import random

random.seed(42)
```

```python
import pandas as pd

data = pd.read_csv(
    '/Users/dima-utyuz/data_analytics/data_analysis/beverage_r.csv',
    delimiter=';',
)
data.head()
```

```python
data.describe()
```

```python
data_shuffled = data.sample(frac=1, random_state=42).reset_index(drop=True)
data_shuffled.head()
```

```python
train_size = int(len(data_shuffled) * 0.66)
test_size = len(data_shuffled) - train_size

X_train = data_shuffled[:train_size]
X_test = data_shuffled[train_size:]

print(len(X_train))
print(len(X_test))
```

```python
parts = [[], [], [], []]

i = 0
for _, row in data_shuffled.iterrows():
    parts[i % 4].append(row)
    i += 1

print(len(parts[0]))
print(len(parts[1]))
print(len(parts[2]))
print(len(parts[3]))
```

```python
groups = [[[], []], [[], []]]

for _, row in data_shuffled.iterrows():
    groups[row['COKE']][row['TAB']].append(row)

sample_a = pd.DataFrame(columns=data.columns)
sample_b = pd.DataFrame(columns=data.columns)

for coke_groups in groups:
    for coke_tab_group in coke_groups:
        sample = pd.DataFrame(coke_tab_group, columns=data.columns)
        m = len(sample) // 2
        sample_a = pd.concat([sample_a, sample[:m]], ignore_index=True)
        sample_b = pd.concat([sample_b, sample[m:]], ignore_index=True)
```
