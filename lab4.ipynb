2.7. Бросаются 4 игральные кости. Найти вероятность того, что на
них выпадет по одинаковому числу очков.

```python
# Аналитическое решение

print(6 / (6 * 6 * 6 * 6))
```

```python
import random

random.seed(42)

# Монте-Карло

exps = 10000000
successes = 0
for _ in range(exps):
    if len({random.randint(1, 6) for _ in range(4)}) == 1:
        successes += 1
print(successes / exps)
```

2.14. Из колоды карт (52 листа) наудачу вынимаются три карты.
Найти вероятность того, что:
- среди них окажется ровно один туз;
- среди них окажется хотя бы один туз;
- это будут тройка, семерка и туз (в любом порядке).

```python
# Аналитическое решение

# 1
print(4 / 52 * (51 - 3) / 51 * (50 - 3) / 50 + (52 - 4) / 52 * 4 / 51 * (50 - 3) / 50 + (52 - 4) / 52 * (51 - 4) / 51 * 4 / 50)

# 2
print(1 - (52 - 4) / 52 * (51 - 4) / 51 * (50 - 4) / 50)

# 3
print(4 / 52 * 4 / 51 * 4 / 50 * 6)
```

```python
cards = []
for i in range(2, 15):
    cards.extend([i, i, i, i])

exps = 10000000
successes_1 = 0
successes_2 = 0
successes_3 = 0
for _ in range(exps):
    hand = random.sample(cards, 3)

    if hand.count(14) == 1:
        successes_1 += 1
    
    if hand.count(14) > 0:
        successes_2 += 1
    
    if set(hand) == {3, 7, 14}:
        successes_3 += 1

print(successes_1 / exps)
print(successes_2 / exps)
print(successes_3 / exps)
```
